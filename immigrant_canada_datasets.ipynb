```python
#How many Austrlian immigrants come in canada in each year
Australian_im = can_im_df.loc["Australia", years]
Australian_im.head(5)
```

```python
#Using above code draw simple plot line
Australian_im.plot()
```

```python
#Change the data types of Australian_im string to int, Also changed data type soh years str to int
Australian_im.index = Australian_im.index.map(int)
#Draw line plot
Australian_im.plot(kind="line", color="red", marker="D")
plt.title("Immigration from Australia", size=20)
plt.ylabel("Number of immigrants")
plt.xlabel("years")
plt.text(2005, 1100, "2013 immigrants")
plt.show()
```

```python
#compare the number of immigrants from Australia , Belgium from 1980 to 2013
Austria_Bel = can_im_df.loc[["Austria", "Belgium"], years]
Austria_Bel.head()
```

```python
# convert country index to column and years column to index:

Austria_Bel = Austria_Bel.transpose()
Austria_Bel.head(5)
```

```python
Austria_Bel = Austria_Bel.index.map(int)
Austria_Bel.plot(kind="line")
plt.title("Immigrants from Austria and Belgium")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")
plt.show()
```
